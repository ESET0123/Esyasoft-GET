# Matplotlib

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
x=[0,2,4,6,8]
y=[0,4,16,36,64]
fig,ax=plt.subplots()
ax.plot(x,y,marker='o', label="Data Points")
ax.set_title("Basic components of Matplotlib Figure")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
arr=np.array([10,20,30])
brr=np.array([5,8,10])
df=pd.DataFrame({"Numbers":arr,"x":brr})
plt.bar(df["x"],df["Numbers"])
plt.show()
```
